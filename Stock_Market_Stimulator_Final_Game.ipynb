```python
import numpy as np
import matplotlib.pyplot as plt
import time
mu = 0.01

sigma = 0.09
start_price = 100
balance = 1000
portfolio = 0
stock_history = []
stock_date = []
buy = False
sell = False
returns = np.random.normal(loc=mu, scale = sigma, size = 1)
price = start_price*(1+returns)
stock_history.append(price)
print("Here are the rules!")
print("This simulator will go for 10 days")
print("It'll say the current price and ask if you want to Buy, Sell or Do Nothing")
print("To do nothing just press Enter")
print("To Buy, press B and then insert the amount you want to buy")
print("To Sell, press S and then insert the amount you want to sell")
print("Yesterday's Price is!: " + str(price))
i = 0
stock_date.append(i)
while i < 400:
  returns = np.random.normal(loc=mu, scale = sigma, size = 1)
  price = price*(1+returns).cumprod()
  print("Today's Price is: " + str(price))
  i += 1
  time.sleep(0.1)
  #decision = input("What are we going to do today?: ")
  decision = 'c'
  if (decision == 'B' or decision == 'b'):
    buy = True
    buy_number = input("How many Shares?: ")
  if (decision == 'S' or decision == 's'):
    sell = True
    sell_number = input("How many Shares?: ")
  if (buy == True):
    buy_order_size = price*int(buy_number)
    if (int(buy_order_size) <= balance):
      balance = balance - price*int(buy_number)
      portfolio = portfolio + int(buy_number)
    else:
      print("Insufficient Funds")
  if (sell == True):
    if(int(sell_number) <= portfolio):
      balance = balance + price*int(sell_number)
      portfolio = portfolio - int(sell_number) 
    else:
      print("Insufficient Funds")
  buy = False
  sell = False
  stock_date.append(i)
  print("The current date is: " + str(stock_date))
  stock_history.append(price)
  print("The current stock history is: " + str(stock_history))
  plt.plot(stock_date, stock_history, label = "Stock Line")
  plt.show()
  print("Your portfolio is: " + str(portfolio))
  print("Your balance is: " + str(balance))
  if (i == 10):
    balance = balance + price*portfolio
    portfolio = portfolio - portfolio
  print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
  
```
